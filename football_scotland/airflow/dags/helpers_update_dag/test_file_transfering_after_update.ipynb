```python
import shutil
import os
import glob

def copy_single_csvs():
    #football_data
    try:
        source_dir_football_data = ''
        destination_dir_football_data = ''
        # Find the single CSV file in the source directory
        csv_files_football_data = glob.glob(os.path.join(source_dir_football_data, "*.csv"))

        # Check if exactly one CSV file exists in the source directory
        if len(csv_files_football_data) != 1:
            print(f"There should be exactly one CSV file in the{source_dir_football_data}  source directory.")
            return

        # Get the file path of the single CSV file
        csv_file_football_data = csv_files_football_data[0]

        # Ensure the destination directory exists
        if not os.path.exists(destination_dir_football_data):
            os.makedirs(destination_dir_football_data)

        # Copy the CSV file to the destination directory
        shutil.copy(csv_file_football_data, destination_dir_football_data)
        print(f"File '{csv_file_football_data}' has been copied to '{destination_dir_football_data}'.")

    except Exception as e:
        print(f"Error occurred while copying the CSV file of football data: {e}")
    
    #fbref_match_reports
    try:
        source_dir_fbref_match_reports= ''
        destination_dir_fbref_match_reports = ''
        # Find the single CSV file in the source directory
        csv_files_fbref_match_reports = glob.glob(os.path.join(source_dir_fbref_match_reports, "*.csv"))

        # Check if exactly one CSV file exists in the source directory
        if len(csv_files_fbref_match_reports) != 1:
            print(f"There should be exactly one CSV file in the{source_dir_fbref_match_reports}  source directory.")
            return

        # Get the file path of the single CSV file
        csv_file_fbref_match_reports= csv_files_fbref_match_reports[0]

        # Ensure the destination directory exists
        if not os.path.exists(destination_dir_fbref_match_reports):
            os.makedirs(destination_dir_fbref_match_reports)

        # Copy the CSV file to the destination directory
        shutil.copy(csv_file_fbref_match_reports, destination_dir_fbref_match_reports)
        print(f"File '{csv_file_fbref_match_reports}' has been copied to '{destination_dir_fbref_match_reports}'.")

    except Exception as e:
        print(f"Error occurred while copying the CSV file of fbref match reports: {e}")


    #fbref_general_stats
    try:
        source_dir_fbref_general_stats= ''
        destination_dir_fbref_general_stats = ''
        # Find the single CSV file in the source directory
        csv_files_fbref_general_stats = glob.glob(os.path.join(source_dir_fbref_general_stats, "*.csv"))

        # Check if exactly one CSV file exists in the source directory
        if len(csv_files_fbref_general_stats) != 1:
            print(f"There should be exactly one CSV file in the{source_dir_fbref_general_stats}  source directory.")
            return

        # Get the file path of the single CSV file
        csv_file_fbref_general_stats= csv_files_fbref_general_stats[0]

        # Ensure the destination directory exists
        if not os.path.exists(destination_dir_fbref_general_stats):
            os.makedirs(destination_dir_fbref_general_stats)

        # Copy the CSV file to the destination directory
        shutil.copy(csv_file_fbref_general_stats, destination_dir_fbref_general_stats)
        print(f"File '{csv_file_fbref_general_stats}' has been copied to '{destination_dir_fbref_general_stats}'.")

    except Exception as e:
        print(f"Error occurred while copying the CSV file of fbref general stats: {e}")

    #sofascore
    try:
        source_dir_sofascore= ''
        destination_dir_sofascore= ''
        # Find the single CSV file in the source directory
        csv_files_sofascore = glob.glob(os.path.join(source_dir_sofascore, "*.csv"))

        # Check if exactly one CSV file exists in the source directory
        if len(csv_files_sofascore) != 1:
            print(f"There should be exactly one CSV file in the{source_dir_sofascore}  source directory.")
            return

        # Get the file path of the single CSV file
        csv_file_sofascore= csv_files_sofascore[0]

        # Ensure the destination directory exists
        if not os.path.exists(destination_dir_sofascore):
            os.makedirs(destination_dir_sofascore)

        # Copy the CSV file to the destination directory
        shutil.copy(csv_file_sofascore, destination_dir_sofascore)
        print(f"File '{csv_file_sofascore}' has been copied to '{destination_dir_sofascore}'.")

    except Exception as e:
        print(f"Error occurred while copying the CSV file of fbref general stats: {e}")



if __name__ == "__main__":
    copy_single_csvs()

```
